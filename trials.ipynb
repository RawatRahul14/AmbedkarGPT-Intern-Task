### RAG Pipeline

```python
model = "mistral"
```

```python
from langchain.llms import Ollama
from langchain.chains import RetrievalQA

llm = Ollama(model=model)

qa = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    chain_type="stuff"
)

print(qa("Why did Ambedkar criticize the Shastras?"))
```
